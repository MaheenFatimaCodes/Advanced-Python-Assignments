### <span style="color:purple">Q.8 (Use the Crime Dataset.)</span>

```python
crime_df=pd.read_csv("C:/Users/DELL/OneDrive/Dataset/crime_data.csv")
```

```python
crime_df
```

#### ===============================================================================

### <span style="color:Green">I) find the aggregations like all moments of business decisions for all columns,value counts.</span>


```python
numeric_cols = crime_df.select_dtypes(include=['float64', 'int64']).columns
```

```python
# Moments of business decision summary table
moments_df = pd.DataFrame({
    'Mean': crime_df[numeric_cols].mean(),
    'Median': crime_df[numeric_cols].median(),
    'Mode': crime_df[numeric_cols].mode().iloc[0],
    'Variance': crime_df[numeric_cols].var(),
    'Std Dev': crime_df[numeric_cols].std(),
    'Skewness': crime_df[numeric_cols].skew(),
    'Kurtosis': crime_df[numeric_cols].kurtosis()
})

print("Moments of Business Decisions:\n", moments_df)
```

```python
# Value counts for categorical/state column
print("\nValue counts for States:\n", crime_df['Unnamed: 0'].value_counts())
```

#### =======================================================================================

### <span style="color:Green">II) do the plottings like plottings like histogram, boxplot, scatterplot, barplot, piechart,dot chart.</span>


```python
fig, axes = plt.subplots(3, 2, figsize=(12, 12))

# 1. Histogram
axes[0, 0].hist(crime_df['Murder'], bins=8, color='skyblue', edgecolor='black')
axes[0, 0].set_title('Histogram (Murder Rate)')

# 2. Boxplot
sns.boxplot(data=crime_df[['Murder', 'Assault', 'Rape']], ax=axes[0, 1])
axes[0, 1].set_title('Boxplot')

# 3. Scatterplot
axes[1, 0].scatter(crime_df['Assault'], crime_df['Murder'], color='red')
axes[1, 0].set_title('Scatterplot: Assault vs Murder')

# 4. Barplot
top5 = crime_df.head(5)
axes[1, 1].bar(top5['Unnamed: 0'], top5['Murder'], color='teal')
axes[1, 1].set_title('Barplot (Murder per State)')

# 5. Piechart
axes[2, 0].pie(top5['Assault'], labels=top5['Unnamed: 0'], autopct='%1.1f%%')
axes[2, 0].set_title('Piechart (Assault Share)')

# 6. Dot Plot
axes[2, 1].plot(top5['UrbanPop'], top5['Unnamed: 0'], 'go')
axes[2, 1].set_title('Dot Plot (UrbanPop %)')

plt.tight_layout()
plt.show()
```

### <span style="color:Green">A) change all the categorical columns into numerical by creating Dummies and using label encoder.</span>


```python
bank_cols = bank_df.select_dtypes(include=['object']).columns
```

```python
# Method 1: One-Hot Encoding (Dummies)
bank_encoded = pd.get_dummies(bank_df, columns=bank_cols, drop_first=True)
```

```python
# Method 2: Label Encoding
le = LabelEncoder()
for col in bank_cols:
    bank_df[col + '_encoded'] = le.fit_transform(bank_df[col])
```

```python
bank_df
```

### <span style="color:Green">B) Rename all the column names in DF</span>


```python
bank_renamed_all = bank_df.copy()
bank_renamed_all.columns = [f"col_{i+1}_{col}" for i, col in enumerate(bank_renamed_all.columns)]
```
