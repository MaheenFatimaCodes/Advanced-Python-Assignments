Q.1) Create an defined function with variable outside of a function, and use it inside the function

```python
# Variable declared outside the function
message = "Welcome to Python Programming!"

# User-defined function
def display_message():
    print(message)

# Function call
display_message()
```

##### Conclusion: In this program, I learned how to create a user-defined function and use a variable declared outside the function. This demonstrates the use of a global variable, which can be accessed inside a function without redefining it.

Q.2) Create an defined function with variable inside a function, with the same name as the global variable.

```python
# Global variable
name = "Maheen"

# User-defined function
def display_name():
    name = "Fatima"   # Local variable with the same name
    print("Inside function:", name)

# Function call
display_name()

# Global variable remains unchanged
print("Outside function:", name)
```

##### Conclusion: In this program, I learned that a local variable can have the same name as a global variable. The local variable is used only inside the function, while the global variable remains unchanged outside the function.
#### ===========================================================================

Q.3) Create an defined function where Global variable and Local variable with same name.

```python
# Global variable
count = 10

# User-defined function
def show_count():
    count = 20   # Local variable with the same name
    print("Local variable:", count)

# Function call
show_count()

# Display global variable
print("Global variable:", count)
```

##### Conclusion: In this program, both the global and local variables have the same name. The local variable is used only inside the function, while the global variable keeps its original value outside the function.
