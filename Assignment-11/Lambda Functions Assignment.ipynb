Q.1)  Write a Python program to sort a list of tuples using Lambda.

subject_marks = [('English', 88), ('Science', 90), ('Maths', 97), ('Social sciences', 82)


```python
# List of tuples
subject_marks = [
    ('English', 88),
    ('Science', 90),
    ('Maths', 97),
    ('Social Sciences', 82)
]

# Sort the list based on marks
sorted_marks = sorted(subject_marks, key=lambda x: x[1])

# Display the sorted list
print("Sorted List:")
print(sorted_marks)
```

##### Conclusion:The program uses a lambda function with the sorted() function to sort a list of tuples based on the second element (marks). The lambda function acts as a key that tells Python to compare the marks instead of the subject names. This makes the code simple, readable, and efficient for sorting data in different ways.
