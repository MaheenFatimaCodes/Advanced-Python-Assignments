Q.1) Write a Python program to get the current time in Python. 
Sample Format :  13:19:49.078205


```python
from datetime import datetime

current_time = datetime.now().time()
print(current_time)
```

##### Conclusion: datetime.now().time() returns the current local time in the format HH:MM:SS.microseconds.

Q.2) Write a Python program to subtract five days from current date. 
Sample Date :
Current Date : 2015-06-22
5 days before Current Date : 2015-06-17


```python
from datetime import datetime, timedelta

current_date = datetime.now().date()
new_date = current_date - timedelta(days=5)

print("Current Date :", current_date)
print("5 days before Current Date :", new_date)
```

##### Conclusion: This program uses the datetime module to get the current date and the timedelta function to subtract 5 days. It helps perform date calculations easily and accurately in Python.

Q.3) Write a Python script to display the various Date Time formats.

a) Current date and time
b) Current year
c) Month of year
d) Week number of the year
e) Weekday of the week
f) Day of year
g) Day of the month
h) Day of week


```python
from datetime import datetime

now = datetime.now()

print("Current date and time :", now)
print("Current year :", now.strftime("%Y"))
print("Month of year :", now.strftime("%m"))
print("Week number of the year :", now.strftime("%U"))
print("Weekday of the week :", now.strftime("%A"))
print("Day of year :", now.strftime("%j"))
print("Day of the month :", now.strftime("%d"))
print("Day of week :", now.strftime("%w"))
```

##### Conclusion: This program uses the datetime module and strftime() function to display different date and time formats such as the current year, month, week number, weekday, and day information. It helps in extracting and formatting date-time values easily in Python.
#### ==================================================================================

Q.4) Write a Python program to convert a string to datetime. 

Sample String : Jan 1 2014 2:43PM
Expected Output : 2014-07-01 14:43:00


```python
from datetime import datetime

date_string = "Jan 1 2014 2:43PM"

date_object = datetime.strptime(date_string, "%b %d %Y %I:%M%p")

print(date_object)
```

##### Conclusion: This program uses the strptime() method of the datetime module to convert a string into a datetime object. It allows Python to interpret date and time strings according to a specified format and convert them into a standard datetime format.

Q.5) Write a Python program to print yesterday, today, tomorrow.

```python
from datetime import date, timedelta

today = date.today()
yesterday = today - timedelta(days=1)
tomorrow = today + timedelta(days=1)

print("Yesterday :", yesterday)
print("Today     :", today)
print("Tomorrow  :", tomorrow)
```

##### Conclusion: This program uses the date and timedelta classes from the datetime module to find yesterday's and tomorrow's dates relative to today. It demonstrates simple date arithmetic in Python.

Q.6) Write a Python program to add 5 seconds with the current time

```python
from datetime import datetime, timedelta

current_time = datetime.now()
new_time = current_time + timedelta(seconds=5)

print("Current Time :", current_time.time())
print("Time after 5 seconds :", new_time.time())
```

##### Conclusion: This program uses datetime.now() to get the current time and timedelta(seconds=5) to add 5 seconds. It demonstrates how to perform time calculations in Python easily.

Q.7)  Write a Python program to get week number. 

Sample Date : 2015, 6, 16

Expected Output : 25


```python
from datetime import date

sample_date = date(2015, 6, 16)

week_number = sample_date.strftime("%U")

print("Week Number:", week_number)
```

##### Conclusion: This program uses the strftime("%U") method to find the week number of a given date. It helps determine which week of the year a particular date belongs to.
#### ==================================================================================

Q.8) Write a Python program to drop microseconds from datetime

```python
from datetime import datetime

current_datetime = datetime.now()

new_datetime = current_datetime.replace(microsecond=0)

print("Original Datetime :", current_datetime)
print("Without Microseconds :", new_datetime)
```

##### Conclusion: This program uses the replace() method with microsecond=0 to remove microseconds from a datetime object. It helps display date and time in a cleaner format without fractional seconds.
#### ==================================================================================

Q.9) Write a Python program to get days between two dates. 

Sample Dates : 2000,2,28, 2001,2,28

Expected Output : 366 days, 0:00:00


```python
from datetime import date

date1 = date(2000, 2, 28)
date2 = date(2001, 2, 28)

difference = date2 - date1

print("Difference:", difference)
```

##### Conclusion: This program uses the date class from the datetime module and subtracts two dates to find the time difference between them. The result is returned as a timedelta object showing the number of days between the dates.

Q.10)Write a Python program to get the number of days of a given month and year.

```python
import calendar

year = 2024
month = 2

days = calendar.monthrange(year, month)[1]

print("Number of days:", days)
```

##### Conclusion: This program uses the calendar module and the monthrange() function to find the number of days in a given month and year. It correctly handles leap years and returns the total days in the specified month.
