Q.1) Which method/inbuilt function should I use to convert String "welcome to the beautiful world of python" to "Welcome To The Beautiful World Of Python".

```python
text = "welcome to the beautiful world of python"
result = text.title()

print(result)
```

##### Conclusion: In Python, the title() method is used to convert the first letter of each word in a string to uppercase, making it useful for proper formatting of sentences and titles.

Q.2) Split a given string on hyphens into several substrings

str1 = Emma-is-a-data-scientist


```python
# Given string
str1 = "Emma-is-a-data-scientist"

# Splitting the string on hyphen (-)
result = str1.split("-")

# Display result
print(result)
```

##### Conclusion: In Python, the split() method is used to divide a string into multiple substrings based on a separator. By using "-" as the delimiter, the string is split at each hyphen and stored as a list of words.

Q.3) Remove empty strings from a list of strings
        
str_list = ["Emma", "Jon", "", "Kelly", None, "Eric", ""]

Expected Output:

Original list of sting
['Emma', 'Jon', '', 'Kelly', None, 'Eric', '']

After removing empty strings
['Emma', 'Jon', 'Kelly', 'Eric']


```python
# Given list
str_list = ["Emma", "Jon", "", "Kelly", None, "Eric", ""]

# Removing empty strings and None values
result = [i for i in str_list if i]

# Display result
print(result)
```

##### Conclusion: In Python, empty strings ("") and None values are considered False in boolean context. By using list comprehension with condition if i, we can easily remove all empty or null values from a list.

Q.4) Create an string and execute the following inbuilt functions on it.
      
- upper, lower, capitalize, split, join, isnumeric, isalpha, isdigit, count, isalnum.


```python
# Create a string
text = "python is 123 awesome"

print("Original String:", text)

# upper()
print("Upper:", text.upper())

# lower()
print("Lower:", text.lower())

# capitalize()
print("Capitalize:", text.capitalize())

# split()
words = text.split()
print("Split:", words)

# join()
joined = "-".join(words)
print("Join:", joined)

# isnumeric()
print("Is Numeric:", text.isnumeric())

# isalpha()
print("Is Alpha:", text.isalpha())

# isdigit()
print("Is Digit:", text.isdigit())

# count()
print("Count of 'o':", text.count("o"))

# isalnum()
print("Is Alnum:", text.isalnum())
```

##### Conclusion: In this program, we learned important string inbuilt functions in Python such as upper(), lower(), capitalize(), split(), join(), count(), and checking functions like isnumeric(), isalpha(), isdigit(), and isalnum(). These functions help in easy manipulation, analysis, and formatting of strings.

Q.7) Find all occurrences of “USA” in given string ignoring the case

str1 = "Welcome to USA. usa awesome, isn't it?"

Expected Outcome:The USA count is: 2


```python
str1 = "Welcome to USA. usa awesome, isn't it?"

# Convert string to lower case and count "usa"
count = str1.lower().count("usa")

print("The USA count is:", count)
```

##### Conclusion: In Python, we can ignore case sensitivity by converting the string to lowercase using lower() and then using the count() method. This helps to accurately find all occurrences of a word regardless of its case.
