Q.1) What is the output of the following addition (+) operator

a = [10, 20]
b = a
b += [30, 40]

print(a)
print(b)


```python
a = [10, 20]
b = a
b += [30, 40]
```

```python
print(a)
```

```python
print(b)
```

##### Answer: a and b both contain [10, 20, 30, 40].
##### Conclusion:Since b = a, both variables refer to the same list object. The += operator modifies the list in place, so the changes made through b are also reflected in a. Therefore, both a and b become: [10, 20, 30, 40].

Q.4) What is the output of the expression  print(-18 // 4) ?

```python
print(-18 // 4) 
```

##### Explaination: -18 / 4 = -4.5, floor value of -4.5 is -5
##### Conclusion: The // operator returns the floor value of the division result. Since -18 ÷ 4 = -4.5, Python rounds down to -5. Therefore, the output is -5.
#### =====================================================================

Q.5) What is the output of print(2 ** 3 ** 2) ?

```python
print(2 ** 3 ** 2) 
```

##### Explaination: 3 ** 2 = 9,2 ** 9 = 512
##### Conclusion: Python evaluates exponentiation (**) from right to left. Therefore, 2 ** 3 ** 2 is calculated as 2 ** (3 ** 2), which equals 512.
