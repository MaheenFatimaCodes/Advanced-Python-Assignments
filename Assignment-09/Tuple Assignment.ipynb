Q.1)  Select which is true for Python tuple

a)	  A tuple maintains the order of items

b)	  A tuple is unordered

c)	  We cannot change the tuple once created

d)	  We can change the tuple once created


##### Correct Answers:
###### a) A tuple maintains the order of items
###### c) We cannot change the tuple once created

##### Conclusion: In Python, tuples are ordered and immutable data structures. This means the elements in a tuple maintain their order, but once created, they cannot be changed. Therefore, options (a) and (c) are correct, while (b) and (d) are incorrect.

Q.2) Access value 20 from the following tuple
       
aTuple = ("Orange", [10, 20, 30], (5, 15, 25))

Expected output: 20


```python
# Given tuple
aTuple = ("Orange", [10, 20, 30], (5, 15, 25))

# Access value 20
result = aTuple[1][1]

# Display result
print(result)
```

##### Conclusion: In Python, we can access elements inside nested data structures using indexing. In this case, the tuple contains a list, so we first access the list using index [1] and then access the value 20 using [1] again.

Q.3) Unpack the following tuple into 4 variables

aTuple = (10, 20, 30, 40)

Expected output:aTuple = (10, 20, 30, 40)

print(a) # should print 10

print(b) # should print 20

print(c) # should print 30

print(d) # should print 40


```python
# Given tuple
aTuple = (10, 20, 30, 40)

# Unpacking tuple
a, b, c, d = aTuple

# Display results
print(a)
print(b)
print(c)
print(d)
```

##### Conclusion: In Python, tuple unpacking allows us to assign each element of a tuple to separate variables in a single line. This makes the code simple, readable, and efficient for handling multiple values.

Q.4) Copy element 44 and 55 from the following tuple into a new tuple

tuple1 = (11, 22, 33, 44, 55, 66)

Expected output:tuple2 = (44, 55)


```python
# Given tuple
tuple1 = (11, 22, 33, 44, 55, 66)

# Copy elements 44 and 55 into a new tuple
tuple2 = tuple1[3:5]

# Display result
print("tuple2 =", tuple2)
```

##### Conclusion: In Python, we can extract specific elements from a tuple using slicing. Since tuples are ordered, we can easily copy required values into a new tuple by specifying index ranges.

Q.5) Counts the number of occurrences of item 50 from a tuple

tuple1 = (50, 10, 60, 70, 50)


```python
# Given tuple
tuple1 = (50, 10, 60, 70, 50)

# Count occurrences of 50
count_50 = tuple1.count(50)

# Display result
print(count_50)
```

##### Conclusion: In Python, we can count how many times an item appears in a tuple using the count() method. It returns the total number of occurrences of the specified value.

Q.6) Modify the first item (22) of a list inside a following tuple to 222

tuple1 = (11, [22, 33], 44, 55)

Expected output:  tuple1 = (11, [222, 33], 44, 55)


```python
# Given tuple
tuple1 = (11, [22, 33], 44, 55)

# Modify first item of the list inside tuple
tuple1[1][0] = 222

# Display result
print(tuple1)
```

##### Conclusion: In Python, tuples are immutable, but if a tuple contains a mutable object like a list, we can modify that inner list. Here, the list inside the tuple is updated by changing its first element.
#### =========================================================================

Q.7) Swap the following two tuples
        
tuple1 = (11, 22)
tuple2 = (99, 88)

Expected output:
tuple1 = (99, 88)
tuple2 = (11, 22)


```python
# Given tuples
tuple1 = (11, 22)
tuple2 = (99, 88)

# Swapping tuples
tuple1, tuple2 = tuple2, tuple1

# Display result
print("tuple1 =", tuple1)
print("tuple2 =", tuple2)
```

##### Conclusion: In Python, we can easily swap two tuples using multiple assignment in a single line. This method is simple, efficient, and does not require a temporary variable.
#### =========================================================================

Q.8) Reverse the following tuple

aTuple = (10, 20, 30, 40, 50)

Expected output:
(50, 40, 30, 20, 10)


```python
# Given tuple
aTuple = (10, 20, 30, 40, 50)

# Reversing the tuple using slicing
reversed_tuple = aTuple[::-1]

# Display result
print(reversed_tuple)
```

##### Conclusion: In Python, we can reverse a tuple using slicing with [::-1]. This creates a new tuple with elements in reverse order without modifying the original tuple.
