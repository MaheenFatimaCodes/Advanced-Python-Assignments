Q.1) Select the right way to create a string literal Ault'Kelly


a)	  str1 = ‘Ault\\’Kelly’
b)	  str1 = ‘Ault\’Kelly’
c)	  str1 = “””Ault’Kelly”””


```python
str1 = 'Ault\\’Kelly'
str1
```

```python
str2 = 'Ault\’Kelly'
str2
```

```python
str3 = """Ault’Kelly"""
str3
```

##### Conclusion:
##### The string Ault'Kelly contains a single quote ('). Using triple double quotes allows Python to store the string without needing an escape character. Therefore, option (c) is a valid way to create the string literal.

Q.9) A string is immutable in Python?
Every time when we modify the string, Python Always create a new String and assign a new string to that variable.

a)	  True
b)	  False


```python
str1 = "Hello"
print(id(str1))

str1 = str1 + " World"
print(id(str1))
```

##### Answer:a
##### Conclusion:Strings in Python are immutable, which means their contents cannot be changed after creation. When a string is modified, Python creates a new string object instead of changing the existing one.
