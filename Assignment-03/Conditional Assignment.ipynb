Q.5)A shop will give discount of 10% if the cost of purchased quantity is more than 1000.
Ask user for quantity
Suppose, one unit will cost 100.
Judge and print total cost for user.


```python
quantity=int(input("Enter Quantity of product:"))

cost=quantity*100

if cost > 1000:
    discount=cost*0.10
    cost=cost-discount

print("Total Cost:",cost)    
```

##### Conclusion: This program calculates the total purchase cost based on the quantity entered by the user. If the total cost exceeds ₹1000, a 10% discount is applied; otherwise, the user pays the full amount.

Q.7)Write a Python program that accepts a string and calculate the number of digits and letters.
Sample Data : Python 3.2
Expected Output :
Letters 6
Digits 2

```python
text = input("Enter a string: ")

letters = 0
digits = 0

for ch in text:
    if ch.isalpha():
        letters += 1
    elif ch.isdigit():
        digits += 1

print("Letters", letters)
print("Digits", digits)
```

##### Conclusion: This program takes a string as input and checks each character. It counts alphabetic characters as letters and numeric characters as digits, then displays the total number of letters and digits in the string.

Q.8)Write a Python program to check whether an alphabet is a vowel or consonant. 
Expected Output:
Input a letter of the alphabet: k                                       
k is a consonant.


```python
letter = input("Input a letter of the alphabet: ")

if letter.lower() in ('a', 'e', 'i', 'o', 'u'):
    print(letter, "is a vowel.")
else:
    print(letter, "is a consonant.")
```

##### Conclusion: This program takes a letter as input and checks whether it is one of the vowels (a, e, i, o, u). If it is, the program prints vowel; otherwise, it prints consonant. The lower() function allows the program to work for both uppercase and lowercase letters.

Q.9)Write a Python program to check a triangle is equilateral, isosceles or scalene. Go to the editor
Note :
An equilateral triangle is a triangle in which all three sides are equal.
A scalene triangle is a triangle that has three unequal sides.
An isosceles triangle is a triangle with (at least) two equal sides.
Expected Output:
Input lengths of the triangle sides:                                    
x: 6                                                                    
y: 8                                                                    
z: 12                                                                   
Scalene triangle 

```python
x = float(input("x: "))
y = float(input("y: "))
z = float(input("z: "))

if x == y == z:
    print("Equilateral triangle")
elif x == y or y == z or x == z:
    print("Isosceles triangle")
else:
    print("Scalene triangle")
```

##### Conclusion: This program compares the three sides of a triangle: If all three sides are equal, it is an Equilateral Triangle.If any two sides are equal, it is an Isosceles Triangle.If all three sides are different, it is a Scalene Triangle.

Q.10)Write an If elif else condition for finding the largest number in 3 numbers.

```python
a = float(input("Enter first number: "))
b = float(input("Enter second number: "))
c = float(input("Enter third number: "))

if a >= b and a >= c:
    print("Largest number is:", a)
elif b >= a and b >= c:
    print("Largest number is:", b)
else:
    print("Largest number is:", c)
```

##### Conclusion: This program compares three numbers using if-elif-else conditions and prints the largest number among them. It checks each number against the other two to determine which one has the greatest value.
