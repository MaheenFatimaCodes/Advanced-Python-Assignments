Q.2)Using while loop accept numbers until the sum of number is less than 100.

```python
total = 0

while total < 100:
    num = int(input("Enter a number: "))
    total += num

print("Sum is:", total)
```

##### Conclusion: This program keeps asking the user to enter numbers using a while loop. It continuously adds the numbers to total until the sum becomes 100 or more, then stops and prints the final sum.

Q.3)Take 10 integers from keyboard using loop and print their average value on the screen.

```python
sum = 0

for i in range(10):
    num = int(input("Enter a number: "))
    sum += num

average = sum / 10
print("Average =", average)
```

##### Conclusion: This program takes 10 integers from the user using a loop, calculates their total sum, and then divides it by 10 to find the average value, which is finally displayed on the screen.

```python
# b.
n = 3

for i in range(1, n + 1):
    print(" " * (n - i) + "*" * (2 * i - 1))

for i in range(n - 1, 0, -1):
    print(" " * (n - i) + "*" * (2 * i - 1))
```

Q.6)Given a list iterate it and display numbers which are divisible by 5 and if you find number greater than 150 stop the loop iteration
list1 = [12, 15, 32, 42, 55, 75, 122, 132, 150, 180, 200]
         
Expected output:
15
55
75
150


```python
list1 = [12, 15, 32, 42, 55, 75, 122, 132, 150, 180, 200]

for num in list1:
    if num > 150:
        break
    if num % 5 == 0:
        print(num)
```

##### Conclusion: This program iterates through the given list and prints numbers that are divisible by 5. The loop stops immediately when a number greater than 150 is encountered, using the break statement.

Q.7)Given a number count the total number of digits in a number
For example, the number is 75869, so the output should be 5.

```python
num = int(input("Enter a number: "))

count = 0

while num != 0:
    num = num // 10
    count += 1

print("Total digits =", count)
```

##### Conclusion: This program counts the number of digits in a given number by repeatedly dividing the number by 10 until it becomes zero, and increasing a counter each time. The final value of the counter gives the total number of digits.

Q.10) Print the following pattern using nested for loop
Expected output:
###### 5 4 3 2 1 
###### 4 3 2 1 
###### 3 2 1 
###### 2 1 
###### 1


```python
for i in range(5, 0, -1):
    for j in range(i, 0, -1):
        print(j, end=" ")
    print()
```

##### Conclusion: This program uses a nested for loop to print a decreasing number pattern. The outer loop controls the starting number of each row, while the inner loop prints numbers in reverse order from that starting value down to 1, forming the required pattern.
