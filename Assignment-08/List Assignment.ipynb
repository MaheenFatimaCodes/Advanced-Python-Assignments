Q.1) Select all the correct options to join two lists.

listOne  =  ['a', 'b', 'c', 'd']

a)	newList = listOne + listTwo

b)	  newList = extend(listOne, listTwo)

Q.2) Create an list of elements and execute the following functions: reverse, append, extend, count,sort.

```python
# Create a list
my_list = [5, 2, 9, 1, 5, 6]

print("Original List:", my_list)

# reverse()
my_list.reverse()
print("After reverse:", my_list)

# append()
my_list.append(10)
print("After append(10):", my_list)

# extend()
my_list.extend([20, 30])
print("After extend([20, 30]):", my_list)

# count()
count_5 = my_list.count(5)
print("Count of 5:", count_5)

# sort()
my_list.sort()
print("After sort:", my_list)
```

##### Conclusion: In this program, we learned important list functions in Python such as reverse(), append(), extend(), count(), and sort(). These functions help us modify, expand, analyze, and organize list data efficiently.

Q.3)  Write a Python program to print a specified list after removing the 0th, 4th and 5th elements.

Sample List : ['Red', 'Green', 'White', 'Black', 'Pink', 'Yellow']

Expected Output : ['Green', 'White', 'Black']

```python
# Sample list
colors = ['Red', 'Green', 'White', 'Black', 'Pink', 'Yellow']

# Removing 0th, 4th, and 5th elements
result = [colors[i] for i in range(len(colors)) if i not in (0, 4, 5)]

# Display result
print("Original List:", colors)
print("Updated List:", result)
```

##### Conclusion: In Python, we can remove specific elements from a list using indexing with conditions. By excluding unwanted index positions, we can create a new list without modifying the original data. This method is simple and useful for selective deletion of elements.

Q.4) What is the output of the following list operation

aList = [10, 20, 30, 40, 50, 60, 70, 80]

print(aList[2:5])
print(aList[:4])
print(aList[3:])


```python
aList = [10, 20, 30, 40, 50, 60, 70, 80]

print(aList[2:5])
print(aList[:4])
print(aList[3:])
```

##### Conclusion: In Python, list slicing is used to access parts of a list using index ranges. The syntax list[start:end] extracts elements from the start index up to (end-1). If the start is omitted, it begins from the first element, and if the end is omitted, it goes till the last element. This helps in easily retrieving required portions of a list.

Q.6) Create an list of elements and slice & dice it.

```python
# Create a list
my_list = [10, 20, 30, 40, 50, 60, 70, 80, 90]

print("Original List:", my_list)

# Slicing examples
print("Elements from index 2 to 5:", my_list[2:6])
print("First 4 elements:", my_list[:4])
print("Elements from index 5 to end:", my_list[5:])
print("Every 2nd element:", my_list[::2])
print("Reverse list:", my_list[::-1])
```

##### Conclusion: In Python, list slicing(slicing,dicing) allows us to extract specific parts of a list using index ranges. It also supports step values for skipping elements or reversing the list. This makes data manipulation easy, flexible, and efficient.

Q.7) Given a Python list you should be able to display Python list in the following order
       
aLsit = [100, 200, 300, 400, 500]

Expected output:[500, 400, 300, 200, 100]

```python
# Given list
aList = [100, 200, 300, 400, 500]

# Reversing the list
reversed_list = aList[::-1]

# Display result
print(reversed_list)
```

##### Conclusion: In Python, we can reverse a list easily using slicing with [::-1]. This method creates a reversed version of the list without modifying the original list. It is a simple and efficient way to change the order of elements.
