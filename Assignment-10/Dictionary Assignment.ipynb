Q.1) Write a Python script to sort (ascending and descending) a dictionary by value.

```python
# Sample dictionary
my_dict = {
    'apple': 50,
    'banana': 20,
    'mango': 80,
    'grapes': 30
}

# Sort in ascending order (by value)
asc_sorted = dict(sorted(my_dict.items(), key=lambda item: item[1]))

# Sort in descending order (by value)
desc_sorted = dict(sorted(my_dict.items(), key=lambda item: item[1], reverse=True))

# Display results
print("Original Dictionary:", my_dict)
print("Ascending Order:", asc_sorted)
print("Descending Order:", desc_sorted)
```

##### Conclusion: Sorting a dictionary by value in Python can be easily done using the sorted() function with a lambda function. By setting reverse=False, we get ascending order, and by setting reverse=True, we get descending order. This helps in efficiently organizing data based on values for better analysis and decision-making.

Q.3) Write a Python script to merge two Python dictionaries.

```python
# First dictionary
dict1 = {1: 10, 2: 20}

# Second dictionary
dict2 = {3: 30, 4: 40}

# Merging dictionaries (Python 3.5+)
merged_dict = {**dict1, **dict2}

# Display result
print("Dictionary 1:", dict1)
print("Dictionary 2:", dict2)
print("Merged Dictionary:", merged_dict)
```

##### Conclusion: Two dictionaries in Python can be merged using the unpacking operator **. This method creates a new dictionary containing all key-value pairs from both dictionaries, making it a simple and efficient way to combine data.

Q.4) Write a Python program to sum all the items in a dictionary.

```python
# Sample dictionary
my_dict = {1: 10, 2: 20, 3: 30, 4: 40}

# Sum of all values in dictionary
total_sum = sum(my_dict.values())

# Display result
print("Dictionary:", my_dict)
print("Sum of all items:", total_sum)
```

##### Conclusion: In Python, we can calculate the sum of all values in a dictionary using the sum() function along with dict.values(). This provides a simple and efficient way to perform aggregation on dictionary data.

Q.5) Write a Python program to combine two dictionary adding values for common keys. 

d1 = {'a': 100, 'b': 200, 'c':300}
d2 = {'a': 300, 'b': 200, 'd':400}

Sample output: Counter({'a': 400, 'b': 400, 'd': 400, 'c': 300})


```python
from collections import Counter

# Given dictionaries
d1 = {'a': 100, 'b': 200, 'c': 300}
d2 = {'a': 300, 'b': 200, 'd': 400}

# Combine dictionaries by adding values of common keys
result = Counter(d1) + Counter(d2)

# Display result
print(result)
```

##### Conclusion: In Python, we can combine two dictionaries and automatically add values of common keys using Counter from the collections module. When two Counter objects are added, values of matching keys are summed, and unique keys are included directly in the result.

Q.7) Write a Python program to count number of items in a dictionary value that is a list.

```python
# Sample dictionary
my_dict = {
    'a': [1, 2, 3],
    'b': [4, 5],
    'c': 10,
    'd': [6, 7, 8, 9]
}

# Count items in list-type values
total_count = 0

for value in my_dict.values():
    if isinstance(value, list):
        total_count += len(value)

print("Dictionary:", my_dict)
print("Total number of items in list values:", total_count)
```

##### Conclusion: In Python, we can count the total number of elements inside list-type values in a dictionary by checking each value using isinstance() and then adding their lengths using len(). This helps in handling mixed-type dictionary data efficiently.
